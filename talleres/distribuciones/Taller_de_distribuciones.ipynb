# Primera Distribución

Distribución exponencial


```python
import numpy as np
from matplotlib import pyplot as plt

rng = np.random.default_rng()
valores =  [0.1,0.5,1.0,4.0,10.0,20.0]

for i in range(len(valores)):
  expo = rng.exponential(valores[i],5000)
  plt.figure(figsize=(20,12))
  plt.subplot(5,4,i+1).set_title("Valor del arreglo: "+str(valores[i]))
  plt.hist(expo)

plt.show()
```

**Conclusión:**  esta distribución es muy usada para la probabilidad y para su cálculo se pide una entrada la cúal es el **parámetro de escala**. Este paŕametro  se recomienda que sea flotante y positivo;  este es el inverso de la tasa de que un evento ocurra en  x tiempo, es decir que a mayor el **parámetro de escala** menor es la tasa de frecuencia y mientrás más pequeño sea el parámetro de escala, mayor es el parámetro de tasa. Por esta razón se puede observar que:
- Cuando el parámetro de escala es menor implica que la tasa de frecuencia es tan grande que las probabilidades de que ocurran son pocas.

- Cuando el parámetro de escala es mayor implica que la tasa de frecuencia es tan pequeña que las probabilidades de que ocurran son muy altas.

# Segunda  Distribución

Distribución binomial


La Distribución binomial cuenta el número de éxitos en una secuencia de *n* ensayos independientes entre si y con una probabilidad fija *p* de ocurrencia de éxito entre los ensayos.

```python
rng = np.random.default_rng()
n =  [1,2,5,10,25]
p =  [0.1,0.3,0.9]

for i in range(len(n)):
  for j in range(len(p)):
    bino = rng.binomial(n[i],p[j],5000)
    plt.figure(figsize=(20,12))
    plt.subplot(5,4,i+1).set_title("Valor de n: "+str(n[i])+" Valor de p: "+str(p[j]))
    plt.hist(bino)

plt.show()
```

**Conclusión:** es posible observar las siguientes relaciones entre p y n:

- Cuando n es bajo y p es bajo entonces el número de exitos en n ensayos es muy alta en cero.

- Cuando n es bajo y p es alto entonces el número de exitos en n ensayos es muy alta en números mayores.

- Cuando n es alto y p es alto entonces el número de exitos en n ensayos es muy bajo con x bastantes pequeñas y grandes, mientras que entre rango hay un punto en x donde la probilidad es máxima, de tal forma que se genera la campana de Gauss.



# Tercera  Distribución






distribucion chisquare



```python
import numpy as np
from matplotlib import pyplot as plt
rng = np.random.default_rng()

plt.figure(figsize=(15,15))

A = [1,5,20,50,70,90,200,400]
for i in range(len(A)):
  chis = rng.chisquare(A[i],1000)
  plt.subplot(4,2,i+1).set_title("df = "+str(A[i]))
  plt.hist(chis)
plt.show()
```

**conclusion:** Al variar el parametro del chisquare lo que varia es el número de muestras que se toman, permitiendo crear intervalos con mayor confianza

# Cuarta  Distribución

distribucion geometrica







```python
plt.figure(figsize=(15,15))

A = [0.1,0.4,0.5,0.6,0.8,0.9]

for i in range(len(A)):
  geo = rng.geometric(A[i],1000)
  plt.subplot(4,2,i+1).set_title("df = "+str(A[i]))
  plt.hist(geo)
plt.show()
```

**conclusion:**
Al variar el parametro de esta distribucion lo que hace es aumentar o disminuir la probabilidad de exito en una muestra individual, entonces entre mas grande el numero dado en el parametro mayor sera la probabilidad de exito, lo mismo en el caso contrario.
